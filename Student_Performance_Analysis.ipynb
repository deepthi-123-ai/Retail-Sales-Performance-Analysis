```python
# ==============================
# Student Performance Analysis
# ==============================

# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("student_performance_dataset (1).csv")

# Display First 5 Rows
print("First 5 Rows:")
print(df.head())

# Dataset Information
print("\nDataset Shape:")
print(df.shape)

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

# Create Average Score Column
df["Average_Score"] = (
    df["Math_Score"] +
    df["Science_Score"] +
    df["English_Score"]
) / 3

# Display Updated Dataset
print("\nDataset with Average Score:")
print(df.head())

# Top 10 Students
top_students = df.sort_values(
    by="Average_Score",
    ascending=False
)

print("\nTop 10 Students:")
print(top_students.head(10))

# Gender-wise Average Score
gender_avg = df.groupby("Gender")["Average_Score"].mean()

print("\nGender-wise Average Scores:")
print(gender_avg)

# ==============================
# Visualization 1
# Gender-wise Average Scores
# ==============================
plt.figure(figsize=(6,4))
gender_avg.plot(kind="bar")
plt.title("Average Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.show()

# ==============================
# Visualization 2
# Study Hours vs Average Score
# ==============================
plt.figure(figsize=(6,4))
plt.scatter(df["Study_Hours"], df["Average_Score"])
plt.title("Study Hours vs Average Score")
plt.xlabel("Study Hours")
plt.ylabel("Average Score")
plt.show()

# ==============================
# Visualization 3
# Attendance vs Average Score
# ==============================
plt.figure(figsize=(6,4))
plt.scatter(df["Attendance"], df["Average_Score"])
plt.title("Attendance vs Average Score")
plt.xlabel("Attendance")
plt.ylabel("Average Score")
plt.show()


# Visualization 4
# Score Distribution

plt.figure(figsize=(6,4))
plt.hist(df["Average_Score"], bins=10)
plt.title("Distribution of Average Scores")
plt.xlabel("Average Score")
plt.ylabel("Number of Students")
plt.show()
```
